# Future Value

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# basic example
r = 0.05
T = 1
pv = 50000
fv = pv * (1+r) **T
print(fv)
```

```python
# simple function
def fv(pv,r,T):
    return round(pv * (1 + r) ** T ,2)
```

```python
fv(pv,r,5)
```

```python
# need a currency symbol?
def fv(pv, r, T, add=0, type=0):
    return '${:,.2f}'.format(pv * (1 + r) ** T)   
```

```python
fv(pv,r,5)
```

```python
ending_balance = np.ones(20) * 50000
for T in range(1,len(ending_balance)):
    ending_balance[T] = round(ending_balance[T] * (1 + r) ** T,2)
ending_balance
```

```python
plt.plot(ending_balance)
plt.title("Ending Balance Over Time")
plt.grid(linestyle="--")
plt.xticks([0,5,10,15,20])
```

```python
ending_balance = np.ones(20) * 50000
for T in range(len(ending_balance)):
    r = np.random.normal(.095, .185)
    if T > 0:
        ending_balance[T] = round(ending_balance[T - 1] * (1 + r))
    else: 
        ending_balance[T] = ending_balance[T]     
ending_balance    
```

```python
plt.plot(ending_balance)
plt.title("Ending Balance Over Time")
plt.grid(linestyle="--")
plt.xticks([0,5,10,15,20])
```
